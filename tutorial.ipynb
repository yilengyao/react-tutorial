### useContext
useContext allows components to share data using scoping

`useContext` is a hook in React that allows you to access the value of a context. Context provides a way to pass data through the component tree without having to pass props down manually at every level, making it especially useful for data that can be considered "global" for a tree of React components.

Here's how `useContext` works:

1. Create a Context: First, you'd create a context using `React.createContext()`:

```
const MyContext = React.createContex(defaultValue);
```

This returns a context object(`MyContext` in this case) that has a `Provider` and a `Consumer` component. But with the `useContext` hook, you often don't need to use the `Comsumer` directly.

2. Providing the Context Value: In your component tree, you'd use the `Provider` component to set the current value of the context:

```
<MyContext.Provider value={someValue}>
    {/* children components can access the value now */}
</MyContext.Provider>
```

3. Accessing the Context Value: Any descendant component of `MyContext.Provider` can now access the context value using the `useContext` hook:

```
const value = useContext(MyContext);
```

In this example, `value` would be set to `someValue` (or `defaultValue` if no `Provider` ancestor exists for this context).

Here's a brief example to illusgtate:

```
import React, { useContext } from 'react';

// 1. Creat a context
const ThemeContext = React.createContext('light');

function DisplayTheme() {
    // 3. Access the context value
    const theme = useContext(ThemeContext);
    return <div>The Current theme is {theme}</div>;
}

function App() {
    return (
        // 2. Provide the context value
        <ThemeContext.Provider value="dark">
            <DisplayTheme />
        </ThemeContext.Provider>
    );
}

export default App;
```

In the example above, the `DisplayTheme` component can access the theme value provided by its ancestor `ThemeContext.Provider` component. The displayed result will be "The current theme is dark".

Remember, one of the primary benefits of using context is to avoid "prop drilling" where you pass down props through many levels of components even if intermediate components don't use them. With context, you can directly access the value you need in any component without passing them down through props.

#### useContext and Redux
While both `useContext` and Redux's store achieve the goal of managing and providing access to state across components, they are conceptually and funcationally different. However, 'useContext' combined with the `useReducer` hook can be used to build a simple state management solution that resembles some aspects of Redux.

Here's a bries breakdown:

1. useContext: It provides a way to share values (data or functions) between a parent component and its descendants without having to pass props explicitly. It's like a global variable scoped to a subtree of components.
2. Redux Store: Redus provides a store, which is a single source of truth for you app's state, along with a dispatch mechanism to change the state based on actions. Redux's design emphasizes predictable state updates using pure reducer functions in response to actions.

You can, in a way mimic Redux's functionality using React's built-in hooks:

1. useReducer: This hook can be seem as  a mini Redux in that it manages state and updates it based on dispathed actions through a reducer function. It provides both the current state and a dispatch funciton to update that state.
2. useContext + useReducer: By combining these 2 hooks, you can create a shared state management solution:

```
const StoreContext = React.createContext();

const initialState = { count: 0 };

function reducer(state, action) {
    switch (action.type) {
        case 'increment':
            return { count: state.count + 1 };
        default:
         return state;
    }
}

function App() {
    const [state, dispatch] = useReducer(reducer, initalState);
    
    return (
        <StoreContet.Provider value={{ state, dispatch }}>
            {/* Child components can now access state and dispatch */}
        </StoreContext.Provider>
    );
}
```

In child components:

```
const { state, dispatch } = useContext(StoreContext);
```

Is this the same as Redux? No, but it shares some similarities. Redux comes with more built-in feature (middleware support, devtools, etc.), better scalability for larger apps, and has a whole ecosystem of tools and utilities around it. But for simpler scenarios, `useContext` and `useReducer` can be combined to build a lightweight state management solution without bringing in Redux.

To answer your specific question: `useContext` alone is not a way to create a store in Redux. But when combined with `useReducer`, it can be used to build a state management pattern that has similarities to Redux. If your application grows in complexity, you might still find value in using Redux or other more extensive state management solutions due to their added features and optimizations.

# Flux
Flux: an application architecture utilizing a unidirectional data flow

Action -> Dispatcher -> Store -> View

The Flux architecture and the MVC(Model-View-Controller) pattern are both design patterns for managing the state and flow of data in software applications, but they differ in their approach and structure.

Flux:
1. Components: Flux applicaitons have 3 main components.
- - Dispatcher: A central coordinator that manages all data flow (actions, and update) to the stores in a Flux application
- - Stores: Holds application state logic, Store contains the application state and logic; they're somewhat like a model in a traditional MVC. However, instead of representing a single database recored, they manage the state of many objects. Unlike a model, you represent the data however it makes sense, unconstrained by resources.
- - Views (React Components): Listen to changes in the Stores and re-render themselves, The user interface, usually React, but Flux doen't requrie React.
- - Actions: rather than update state directly, flux apps modify their state by creating actions to modify them.
2. Data Flow: Flux enforces a single unidirectional data flow. Actions are dispatched, stores handle the aciton and change their state, and views re-render.
3. Actions: These are plain objectse describing user interactions, API calls, etc., which are sent to the dispatcher
4. Usage: Disveloped by Facebook alongside react, through it can be used with other libraries/frameworks

MVC:
1. Components:
- - Model: Represents data and the business rules of the applicaiton
- - View: Displays the data, the UI of the application
- - Controller: Acts as an interface betwween Model and View.
2. DataFlow: In MVC, data can flow in multiple directions. The Controller can update the View, the View can request data from the Model, and the Controller can also update the Model.
3. Origin: MVC is a classical design pattern used in many object-oriented languages and frameworks.

#### Model
The data for your application. Usually a noun like User, Account, or Post, for example. Your model should at least have the basic methods to manipulate associated data. In the most abstract sense, the model represents raw data or knowledge. It's where the data intersects with your application code. For example the database might store several properties like access-Sopes, authenticated, and so on. But the model will be able to use this data for a method like isAllowedAccessForResource() on it that will act on the underlying data to model. The model is where raw data converts with your application code.

#### View
A representation of your model. The view is often the user interface itself. The view shouldn't have any logic in ti that isn't related to presentations of the data. For front-end frameworks, this would generally mean that a particular view was directly associated with a resource and would have CRUD (create, read, update, delete) actions associated with it. This isn't how front-end applications are always built anymore.

#### Controller
Controllers are the "glue" that binds the model and view together. Controllers should usually be only glue and not much more (for example, they shouldn't have complex view or database logic in them). You should genrally expect controllers to have far less ability to mutate data than the models they interact with.

Differences:
1. Data Flow:
- - Flux: Unidirectional. The data flows in a single direction: Action -> Dispatcher -> Store -> View
- - MVC: Bidirection. Data can flow in multiple directions, and components can communicate back and forth.
2. State Management:
- - Flux: State is typically kept in Stores
- - MVC: State is in the Model
3. Decision MAking:
- - Flux: Logic and decisions are often made in the Stores
- - MVC: The Controller usually make decisions, handling user input, and updating the Model and View
4. Complexity:
- - Flux: While the unidirectional data flow can simplify complex applicaitons, especially when state changes frequently, it might seem verbose for simple applicaitons.
- - MVC: For applications where the relationship between data, representations, and actions is straightforword. MVC can be simpler. But, in complex applications with interdependent data changes, MVC can become messy.

In conclusion, when both Flux and MVC aim to address challenges in managing application state behavior, they offer different solutions. Flux emphasizes a strict unidirectional data flow to avoid unpredictable behavior, especially in large applicaitons, MVC, being an older and more general pattern, has bidrectional data flow and can be found invarious frameworks across different programming languages.

Yes, that's a concise way to put one of the motivations behind the Flux architecture.

Flux was designed with the specific aim of avoiding the cascading effects and potential race conditions that can occur in traditional MVC architectures, especially in complex applications. By enforcing a unidirectional data flow and having a central dispatcher, Flux ensures that updates to the state do not overlap or cause unpredictable behavior.

In a typical MVC setup, multiple views can update a model directly, and changes in the model can then cascade to other parts of the system, potentially leading to race conditions. In Flux, the dispatcher processes actions one at a time, in the order they are triggered, which simplifies the flow and prevents race conditions.

So, yes, you can say that Flux's design inherently "blocks" to prevent the race conditions and cascading updates that can be encountered in MVC patterns.

# Redux
It is an Architectural concept to manage the application state predictably by using a centralize store for state for the entire application.

Redux is a predictable state container for JavaScript apps.

Redux is a predictable state container for JavaScript application. while it's often used with React, it can be used with any other JavaScript framework or library. It was inspired by Facebook's Flux architecture and the functional programming language Elm. Redux makes it easier to manage and manipulate state in web applications, particularly for large applications with complex state management needs.

Here's a breakdown of Redux:

1. Single Source of Truth: Redux uses a single store to keep the entire application's state. This makes it easier to debug and inspect.
2. You cannot make a mutable update to a state. All updates are dne immutably.
3. State is Read-Only: The only way to change the state is by dispatching an action. An action is a plain JavaScript object describing what happened (e.g, `{type: 'ADD_TODO', text: 'Go to the gym' }`)
4. Changes are Made with Pure Functions: In order to specify how the state tree is transformed by actions, you will use pure reducers. Reducers are functions that take the previous state and an action as arguments and return the next state.
5. The code logic that updates the State immutably based on actions is called a Reducer. We wouldn't not be updating the state directly, the reducer will be cloning/copy the state and then do an update on the clone and then return the copied object.
5. The UI subscribed for a state update, gets the updated state.

Core Concepts:
- Actions: They are payloads of information that send data from your application to your store. They are dispatched by the application based on user interactions, network responses etc.
- Reducers: Actions describing the fact that something happened, but don't specify how the applcation's state changes in response. this is the job of reducers. they take the current state and an action, and return the next state.
- Store: The object that brings actions and reducers together. The stre has the following respponsibilities:
- - Holds the application state
- - Allows access to state via `getState()`
- - Allows state to be updated via `dispatch(action)`
- - Registers listeners via `subscribe(listener)`
- Middleware: Provides a 3rd part extention point between dispatching an action and the moment it reaches the reducer. It's often used for loggin, crash reporting, asynchronous API calls, etc.

How Redux differs from Flux:

While Flux and Redux have similarites (like a unidirectional data flow), they have key differences:

- Redux doesn't have a dispatcher. The role of the dispatcher is incorporated into the store
- Redux has a single store, where Flux promotes multiple stores: Rather than locate state information in multiple stores across the app, Redux apps keep everything in 1 place. In Flux, you can have many different stores. Redux breaks from this and enforces a single global store.
- Redux introduces reducers: Reducers are a more immutable approach to mutation. In Redux, state is changed in a predictable and deterministic way, one part of state at a time, and only in one place (the global state)
- In Redux the state is immutable. when the state changes a new state object is created.
- Redux introduces middleware: Beacuase actions and data flow in a unidirectional way, you can add middleware to your Redux app and inject custom behavior as data is updated
- Redux actions are decoupled from store: Actions-creators don't dispatch anything to the store; instead, they return action object that a centrual dispatcher uses.

Redux gained popularity due to its simplicity, especially whe nused with React. Its ecosystem includes tools like Redux Toolkit (to simplify common Redux use cases) and Redux Saga and Redux Thunk (to handle side effects).

### Middleware

Middleware works by letting you hook into a cycle or precess of some kind in a composable way, meaning that you can create and use multiple middleware functionalities independent of each other in a single project. Thunk: the word "thunk" is a programming term that means "a piece of code that does some delayed work".

`redux-thunk` is a middleware library to enable asynchronous action creation in your application.

redux-thunk works by wrapping the dispatch method of a store so that it will handle dispatching something other than plain object (like Promises, an API dealing with asynchronous flows). ### Reducer
Reducers determine how state should change.

Reducers are pure functions that receive the previous state and an action as arguments and return the next state.

Pure functions: functions that given the same input they will produce the same associate output everytime.

To handle the incoming actions, you need to set up reducers. Actions are just ways to describe that something happened and specify some info about what happened. The job of reducers is to specify how the store state should change in response to these actions.

Setting up reducers to respond to incoming actions, reducers usually use a switch statement to match incoming action types to make updates to state. They return a new copy of the state (not the same version with changes) that will then be used to update the store. Reducers also perform catch-all behavior to ensure that unknown actions just return the existing state.

#### useSelector
`useSelector` is a hook provided by the `react-redux` library, which allows you to extract data from the Redux store state in function components. Before hooks were introduced in React, the common way to access the Redux store state is a component was by using the `connect` higher-order component. With the introduction of hooks, `useSelector` provides a more concise and readable way to access the store state in function components.

How `useSelector` works:

1. Accessing the Redux Store State:
`useSelector` takes a selector function as its argument. This selector function receives the entire Redux store state as its only parameter and returns the part of the state that you want to access.

```
const data = useSelector(state => state.someSliceOfState);
```

2. Re-rendering:
The component will re-render any time the value returned by the selector function changes (using a strict `===` comparison by default). This means if you extract a piece of state using `useSelector` and that specific piece of state change in the Redux store, your component will re-render.

3. Optimizing:
If you need to perform more complex computations or derived data from the state, you can use the `createSelector` function from the `reselect` library to memoize the results, ensuring that the component only re-renders when the derived data actually changes.

Example:

Let's say you have a Redux state that looks like this:

```
{
    user: {
        name: 'John',
        age: 30
   },
   posts: [...]
}
```

If you want to access the `user` data in a function component, you can use the `useSelector` hook like this:

```
import { useSelector } from 'react-redux';

function UserProfile() {

    const user = useSelector(state => state.user);
    
    return (
        <div>
            <h1>{user.name}</h1>
            <p>Age: {user.age}</p>
        </div>
    );
}
```

In this example, the `UserProfile` component will re-render whenever the `user` slice of the state changes.

In summary, `useSeector` provides a way to access the Redux store state in function components, leading to a clener and more readable code compared to the older `connect` approach.

## State Normalization
- Recommended
- No duplication of data
- Creates an ID lookup, keeping data in lookup table by itemid

State normalization is a crucial concept in Redux and is often used to enhance the performance and maintainability of applications. This concepts evolves around the organization of data in the Redux store. Rather than storing data in a nested or hierarchical format, which can lead to complex updatesand data reedundancy, state normalization flattens the data structure and ensures a consistent, predictable format.

Principles of State Normalization:
1. Unique IDs: Every entity (eg, user, post, comment) in the state should have a unique identifier. Even if the data comes from different database or sources, within the Redux store, each item's identifier must be unique.
2. Flat Structure: Entitites are stored in an object or a "table" that holds all entities of the same type. Instead of deeply nested structure, references (usually IDs) are used to point from 1 entity to another, It's akin to the relationships between tables in a database.
3. No Data Duplication: Each entity is stored once in the state. When other parts of the state need to refer to that entity, they use the entity's ID. This approach avoids the complications of updating data in multiple places.

Example:
Consider a blog application with a typical nested structure:
```
// Non-normalized state
{
    posts: [
        {
            id: 1,
            title: 'Post title',
            comments: [
                {
                    id: 1,
                    content: 'Comment here',
                 },
            // more comments...
            ],
        },
        // more posts...
    ]
}
```

Apply normalization
```
// Normalized state
{
    posts: {
        byId: {
            '1': {
                 id: 1,
                 title: 'Post title',
                 comments: [1, 2] // refers to the IDs of the comments
             },
             // more posts...
        },
        allIds: [1, 2, 3] // list of all post IDs
    },
    comments: {
        byId: {
            '1': {
                id: 1,
                content: 'Comment here'
            },
            // more comments...
        },
        allIds: [1, 2, 3] // list of all comment IDs
    }
}
```

Benefits of State normalization:
1. Consistency: By referring to the same object in memory, you ensure that changes in one place reflect throughout the store. Thers's a single "source of truth."
2. Performance: It's easier to make quick, shallow comparisons of data changes when you're dealing with flat data. This aspect is particularly advantageous when you're optimizing React componenents for unnecessary renders.
3. Simplified Reducers: Reducers become simpler because they don't need to deal with complex nested data. This characteristic makes the logic more predictable and easier to debug.
4. Easier Updates: With normalization, you avoid deep cloning of objects and arrays when a piece of nested data changes. You just update the reference, making state updates faster and more memory efficient.

Conclusion:

Normalization makes the redux store more than a client-side database with a consistent predictable strucutre, making the logic for reading from and writing to the store much more straightforwared. It is a key principle, especially in applications with a large, complex, or frequently changing state.
